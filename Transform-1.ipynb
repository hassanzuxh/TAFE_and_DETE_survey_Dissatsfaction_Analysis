from review i can see that the institute-tenure column in the TAFE survey has inconsistent date 
- to handle this; i will have to extract the the first string value as int(year) from the data in institute_tenure and save it in "years_in_institute" column

```python
import numpy as np
import re

TafeClean["years_in_institute"]=np.nan
# function to extract years
def extract_years(value):
    try:
        # i will extract all numerical values using numerical expression.
        parts=re.findall(r'\d+', str(value))
        # if there are years return the first one
        years=[int(part) for part in parts if part]
        # if there are no years return NaN
        return years[0] if years else np.nan
    except:
        return np.nan
# i will apply the function to the "institute _tenure" column and store the result in "years in institute"
TafeClean["years_in_institute"]=TafeClean["institute_tenure"].apply(extract_years)
```

```python
# Extracting the numerical value of each record to determine the age
import numpy as np
import re

combined_survey["age_formatted"]=np.nan
# function to extract years
def extract_age(value):
    try:
        # i will extract all numerical values using numerical expression.
        parts=re.findall(r'\d+', str(value))
        # if there are ages return the first one
        age=[int(part) for part in parts if part]
        # if there are no years return NaN
        return age[0] if age else np.nan
    except:
        return np.nan
# i will apply the function to the "institute _tenure" column and store the result in "years in institute"
combined_survey["age_formatted"]=combined_survey["age"].apply(extract_age)
```

Now i have to change the ages into suitable age ranges for easier analysis
 - i will  group my age_ranges based on the previous ranges used
   - 20 or younger
   - 21 - 25
   - 26 - 30
   - 31 - 35
   - 36 - 40
   - 41 - 50
   - 51 - 60
   - 60 or older

```python
combined_survey["age"].value_counts(dropna=False).sort_index()
```

```python
def age_ranges(values):
    if values < 21:
        return "20 or younger"
    elif 21 <= values < 26:
        return "21-25"
    elif 26 <= values < 31:
        return "26-30"
    elif 31 <= values < 36:
        return "31-35"
    elif 36 <= values < 41:
        return "36-40"
    elif 41 <=  values < 51:
        return "41-50"
    elif 51 <= values < 61:
        return "51-60"
    elif pd.isnull(values):
        return np.nan
    else:
        return "61 or older"

# Assigning the function to age column and storing it in a new created column called  age_ranges column
combined_survey["age_ranges"]=combined_survey["age"].apply(age_ranges)

    
```

    Now tackling the institute_tenure.
- i have to categorize the service years to make the filtering of the column easier
- i will create a new column that will hold the categorized value of each employee's service year in institute

```python
# Observing the institute_tenure column
combined_survey["institute_tenure"].value_counts(dropna=False).sort_values()
```

```python
def service_years(values):
    if values < 3:
        return "Novice"
    elif 3 <= values < 7:
        return "Junior"
    elif 7 <= values < 11:
        return "Intermediate"
    elif pd.isnull(values):
        return np.nan
    else:
        return "senior"

combined_survey["tenure_category"]=combined_survey["institute_tenure"].apply(service_years)
```

                                                        Data Visualization

The Dataframe is now ready for further Analysis and visualization that will answer our various questions and enhance our data-driven decison making
        - i will make use of a python library called matplotlib for my pivot table analysis and graphical visualization

1. To answer the question of whether employees who worked for the institutes for a short or long period of time resignned due to some kind of dissatisfaction

Pivot Table Analysis

```python
dissatisfaction_pivot1=combined_survey.pivot_table(index="tenure_category", values="general_dissatisfaction")
dissatisfaction_pivot1.sort_values("general_dissatisfaction")
```

Visual Analysis

```python
import matplotlib.pyplot as plt
sorted_dissatisfaction_pivot1= dissatisfaction_pivot1.sort_values("general_dissatisfaction")
a=sorted_dissatisfaction_pivot1.plot(kind='bar', stacked=True, rot=0)
#function to put the data label on top of the bar
for p in a.patches:
    width=p.get_width()
    height=p.get_height()
    x,y=p.get_xy()
    a.annotate(f"{height:.2f}",(x + width / 2, y + height),ha="center" , va="center")
plt.title("Dissatisfaction by Tenure_Category")
plt.xlabel("tenure_category")
plt.ylabel("general_dissatisfaction")
plt.legend(title="Dissatisfied", loc='best')
a.yaxis.set_visible(False)
plt.show()
```

- Analysis of Employee Dissatisfaction Based on period of time spent in the institute
    - the graphical representation above illustrates a clear correlation between service_tenure at either of the institutes and the likelihood of resigning due to dissatisfaction.
        - less thans 3 years: employees with less than 3 years of service_tenure cite dissatisfaction in 30% of resignations.
        - 4-7 years: Dissatisfaction slightly increases to 34% for employees with 4-7 years of service_tenure.
        - 7+ years: significantly , employees with 7 or more years of service_tenure report dissatisfaction in approximately 50% of resignations.

    This suggests a progressive rise in dissatisfaction as employees extends there stay in the institution .

2. To answer the question of whether younger or older employees resigned due to some kind of dissatisfaction

```python
dissatisfaction_pivot=combined_survey.pivot_table(index="age_ranges", values="general_dissatisfaction")
dissatisfaction_pivot.sort_values("general_dissatisfaction")
```
